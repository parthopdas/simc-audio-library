```python
import json
```

```python
with open('./youtube-api/videos_list_yutt.json') as f:
    videos = json.loads(f.read())

video_ids = set([x['id'] for x in videos])
```

```python
with open('./youtube-api/playlistitems_list_siri.json') as f:
    playlistitems_list_siri = json.loads(f.read())
```

```python
video_ids_covered = []

for playlist in playlistitems_list_siri.values():
    for video in playlist:
        video_ids_covered.append(video['contentDetails']['videoId'])        
```

```python
print(len(video_ids_covered), len(set(video_ids_covered)))
```

```python
missing = set(video_ids) - set(video_ids_covered)
len(missing)
```

```python
for v in videos:
    if v['id'] in missing:
        print(v['snippet']['publishedAt'], '\t', v['snippet']['title'])
```
